```python
import numpy as np
import matplotlib.pyplot as plt
```

Función Par: 

Es simétrica al eje vertical (eje Y). Ej: Una parabola

f (-x) = f(x)

```python
def f(x):
  return x**2

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
def f(x):
  return -x**2

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color="g")
ax.axvline(x=0, color="r")
```

Función Impar:

es simétrica al eje horizontal (eje X). Por ejemplo, una función cúbica es impar.

f (-x) = -f(x)

```python
def f(x):
  return -x**3

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
def f(x):
  return x**3

N = 1000

x = np.linspace(-10,10,N)

y = - f(x)

plt.plot(x,y)
plt.grid()
```

Función acotada: 

Una función es acotada si su codominio (también conocido como rango o imagen) se encuentra entre dos valores, es decir, está acotado. Esta definición se define como que hay un número m que para todo valor del dominio de la función se cumple que:

-m <= f(x) <= m

Por ejemplo, la función seno o coseno están acotadas en el intervalo [-1, 1] dentro de su co-dominio.


```python
def f(x):
  return np.sin(x)

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
```

# Funciones monótonas

Estas funciones son útiles de reconocer o analizar debido a que nos permiten saber si una función crece o decrece en alguno de sus intervalos. Que algo sea monótono significa que no tiene variaciones. Entonces las funciones monótonas son aquellas que dentro de un intervalo I, perteneciente a los números reales, cumple alguna de estas propiedades:

1. La función es monótona y estrictamente creciente:
“si para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea menor a f(x2)”. En palabras mucho más sencillas, lo que nos dice esta definición es que x1 siempre tiene que ser menor que x2 en nuestro intervalo I, y que al evaluar x2 en la función el resultado de esto siempre será mayor que si evaluamos la función en x1. Para las siguientes tres definiciones restantes no cambia mucho la forma en la que se interpretan.

2. La función es monótona y estrictamente decreciente:
“si para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea mayor a f(x2)” -> Es decir decrece el rango o imagen

3. La función es monótona y creciente:
“si para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea menor **o igual** a f(x2)”

4. La función es monótona y decreciente:
“si para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea mayor **o igual** a f(x2)”

# Las funciones periódicas
son aquellas que se repiten cada cierto periodo, este periodo se denomina con la letra T. La relación que debe cumplir la función para ser periódica es la siguiente.

Por ejemplo, la función seno y coseno son funciones periódicas con un periodo T = 2π. Es decir que si nosotros calculamos f(x) y calculamos f(x + 2π) en la función seno el valor que nos den ambas expresiones es el mismo.

```python
def f(x):
  return np.cos(x) #Función Coseno

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
def f(x):
  return np.cos(x) + 2*np.pi #Función Coseno + 2.Pi

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
```

# Funciones cóncavas y convexas

La forma de demostrar la concavidad de una función se puede hacer a través del análisis de derivadas consecutivas, pero aún no llegamos a eso, así que no te preocupes. A continuación, te dejo una forma súper intuitiva de ver si una función es cóncava o convexa.

Se dice que una función dentro de un intervalo es convexa si la función “abre hacia arriba”. Es decir si se ve la siguiente manera:

```python
def f(x):
  return x**2 # Función Convexa

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
plt.axhline(y=0,color="r")
plt.axvline(x=0,color="r")
```

Ahora, ¿qué sería una función cóncava? Pues así es, lo contrario de una convexa. Se dice que una función dentro de un intervalo es concava si la función “abre hacia abajo”. Es decir si se ve la siguiente manera:

```python
def f(x):
  return -x**2 # Función Concava

N = 1000

x = np.linspace(-10,10,N)

y = f(x)

plt.plot(x,y)
plt.grid()
plt.axhline(y=0,color="r")
plt.axvline(x=0,color="r")
```
